```python
import ee
import geemap
```

```python
ee.Authenticate()
ee.Initialize(project="your-ee-project")
```

```python
m = geemap.Map()

m.setCenter(13.1921, 55.7069, 10)
dswx_s1_collection = ee.ImageCollection("OPERA/DSWX/L3_V1/S1").filterDate(
    "2026-04-01", "2026-06-01"
)


def func_roq(image):
    wtr = image.select("BWTR_Binary_water")
    return wtr.updateMask(wtr.eq(1))


masked_collection = dswx_s1_collection.map(func_roq)

dswx_s1 = masked_collection.reduce(ee.Reducer.sum()).rename("BWTR_Binary_water")

wtr_class_values = [0, 1]

wtr_palette = ["ffffff", "ffbbbb", "0000ff"]

wtr_band = dswx_s1.select("BWTR_Binary_water")
m.addLayer(
    wtr_band, {"min": 0, "max": 10, "palette": wtr_palette}, "Water Classification"
)

m
```
